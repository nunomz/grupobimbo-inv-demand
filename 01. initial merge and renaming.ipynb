import csvs to dfs

```python
df = pd.read_csv("D:/onuno/Downloads/kaggle/final_df.csv",index_col=0)
df_producto_tabla = pd.read_csv("D:/onuno/Downloads/kaggle/producto_tabla.csv")
df_cliente_tabla = pd.read_csv("D:/onuno/Downloads/kaggle/cliente_tabla.csv")
df_town_state = pd.read_csv("D:/onuno/Downloads/kaggle/town_state.csv")
```

merge dataframes

```python
df_merged = df.merge(df_producto_tabla,how='left',on='Producto_ID')
df_merged = df_merged.merge(df_town_state,how='left',on='Agencia_ID')
df_merged = df_merged.merge(df_cliente_tabla,how='left',on='Cliente_ID')
df_merged
```
